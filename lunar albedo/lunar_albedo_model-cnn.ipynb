# Neural Networks

```python
from tensorflow import keras
from keras import layers
```

```python
def padder(dataset,padding_size):
    x,y=dataset.shape
    row_vector=np.zeros((padding_size,x))
    column_vector=np.zeros((y+2*padding_size,padding_size))
    dataset1=np.vstack((row_vector,dataset,row_vector))
    dataset2=np.hstack((column_vector,dataset1,column_vector))
    return dataset2
```

```python
padded_Fe_features_train=padder(Fe_features_train,2)
padded_K_features_train=padder(K_features_train,2)
padded_Th_features_train=padder(Th_features_train,2)
padded_Ti_features_train=padder(Ti_features_train,2)
```

```python
padded_Fe_features_train.shape
```

```python
img_hg,img_wd=padded_Fe_features_train.shape
x_data=[]
y_data=[]
filter_size=5
for i in range(2,img_hg-2):
    for j in range(2,img_wd-2):
        k=i-2
        l=j-2
        fe_patch=padded_Fe_features_train[k:k+filter_size,l:l+filter_size]
        k_patch=padded_K_features_train[k:k+filter_size,l:l+filter_size]
        Th_patch=padded_Th_features_train[k:k+filter_size,l:l+filter_size]
        Ti_patch=padded_Ti_features_train[k:k+filter_size,l:l+filter_size]
        stack_patch=np.dstack((fe_patch,k_patch,Th_patch,Ti_patch))
        x_data.append(stack_patch)
        lunar_pixal=Y_label_train[k,l]
        y_data.append(lunar_pixal)
x_data_array=np.asarray(x_data)
y_data_array=np.array([y_data])
        
```

```python
x_data_array.shape
```

```python
y_data_array.shape
```

```python
y_out=y_data_array.T
```

## Model

```python
input_shape=(5,5,4)
model=keras.Sequential()
model.add(layers.Input(shape=input_shape))
model.add(layers.Conv2D(12, kernel_size=(3, 3), activation='relu', padding='same'))
model.add(layers.Conv2D(24, kernel_size=(3, 3), activation='relu', padding='same'))
model.add(layers.Conv2D(48, kernel_size=(3, 3), activation='relu', padding='same'))
model.add(keras.layers.BatchNormalization())
model.add(keras.layers.Dropout(0.15))
model.add(keras.layers.Activation('relu'))
model.add(layers.Flatten())
model.add(layers.Dense(50,activation='sigmoid'))
model.add(layers.Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
adam = keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
model.compile(loss=keras.losses.mean_squared_error, optimizer=adam,metrics=[keras.metrics.MeanSquaredError(),'acc'])
```

```python
history=model.fit(x_data_array, y_out, batch_size=32, epochs=150,validation_split=0.1)
```

##  Metric for results

```python
import math
import numpy as np

def calculate_psnr(img1, img2):
    # img1 and img2 have range [0, 255]
    img1 = img1.astype(np.float64)
    img2 = img2.astype(np.float64)
    mse = np.mean((img1 - img2)**2)
    if mse == 0:
        return float('inf')
    return 20 * math.log10(1.0 / math.sqrt(mse))
```

```python
import math
import numpy as np
import cv2

def ssim(img1, img2):
    C1 = (0.01 * 1)**2
    C2 = (0.03 * 1)**2

    img1 = img1.astype(np.float64)
    img2 = img2.astype(np.float64)
    kernel = cv2.getGaussianKernel(11, 1.5)
    window = np.outer(kernel, kernel.transpose())

    mu1 = cv2.filter2D(img1, -1, window)[5:-5, 5:-5]  # valid
    mu2 = cv2.filter2D(img2, -1, window)[5:-5, 5:-5]
    mu1_sq = mu1**2
    mu2_sq = mu2**2
    mu1_mu2 = mu1 * mu2
    sigma1_sq = cv2.filter2D(img1**2, -1, window)[5:-5, 5:-5] - mu1_sq
    sigma2_sq = cv2.filter2D(img2**2, -1, window)[5:-5, 5:-5] - mu2_sq
    sigma12 = cv2.filter2D(img1 * img2, -1, window)[5:-5, 5:-5] - mu1_mu2

    ssim_map = ((2 * mu1_mu2 + C1) * (2 * sigma12 + C2)) / ((mu1_sq + mu2_sq + C1) *
                                                            (sigma1_sq + sigma2_sq + C2))
    return ssim_map.mean()


def calculate_ssim(img1, img2):
    '''calculate SSIM
    the same outputs as MATLAB's
    img1, img2: [0, 1]
    '''
    if not img1.shape == img2.shape:
        raise ValueError('Input images must have the same dimensions.')
    if img1.ndim == 2:
        return ssim(img1, img2)
    elif img1.ndim == 3:
        if img1.shape[2] == 3:
            ssims = []
            for i in range(3):
                ssims.append(ssim(img1, img2))
            return np.array(ssims).mean()
        elif img1.shape[2] == 1:
            return ssim(np.squeeze(img1), np.squeeze(img2))
    else:
        raise ValueError('Wrong input image dimensions.')
```

```python
psnr_value=calculate_psnr(Y_label_test,pred_right)
print('psnr value:'+str(psnr_value))
```

```python
ssim_value=calculate_ssim(Y_label_test,pred_right)
print('ssim value:'+str(ssim_value))
```
