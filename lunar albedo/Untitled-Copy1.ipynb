## data preprocessing

```python
# now splitting the label dataset into training set and test set........
```

```python
#function for data splitting:
def data_splitter(data):
    F1=data.iloc[:,0:360]
    F2=data.iloc[:,360:720]
    return F1,F2
```

```python
Y_label_train,Y_label_test=data_splitter(albedo_map_data)
```

```python
# checking that dataset splitting correctly works by plotting graphs
```

```python
plt.subplots(figsize=(8,7))
sns.heatmap(Y_label_train,xticklabels=False,yticklabels=False,cmap='gray')
```

```python
plt.subplots(figsize=(8,7))
sns.heatmap(Y_label_test,xticklabels=False,yticklabels=False,cmap='gray')
```

```python
# now splitting feature training and test sets and converting in numpy array....................
```

```python
Fe_features_train,Fe_features_test=data_splitter(Fe_map_data)
```

```python
K_features_train,K_features_test=data_splitter(K_map_data)
```

```python
Th_features_train,Th_features_test=data_splitter(Th_map_data)
```

```python
Ti_features_train,Ti_features_test=data_splitter(Ti_map_data)
```

```python
#  now, in order to make a feature set, we have to merge all the different chemical datasets into one with proper positon
# so that we can use them to train and predict.
```

```python
# way of merging different chemical datasets: first we will convert the all chemical dataset into a column vector after that
# we will merge all column vector column-wise.
# for label data set, we will also convert it into column vector.
```

```python

def column_converter(data):
    data=StandardScaler().fit_transform(data)
    column_arr=data.reshape(-1,1,order='F') # here order='F' specifies that it will read column vice
    return column_arr
```

```python
Fe_column_train=column_converter(Fe_features_train)
```

```python
Fe_column_train.shape
```

```python
K_column_train=column_converter(K_features_train)

Th_column_train=column_converter(Th_features_train)

Ti_column_train=column_converter(Ti_features_train)
```

```python
# now all the data is converted to column vector now we will add them.
```

```python
X_train=np.concatenate((Fe_column_train,K_column_train,Th_column_train,Ti_column_train),axis=1)
```

```python
X_train.shape
```

```python
# now we have our feature set ready,now converting label set.
```

```python
Y_train=Y_label_train.values.reshape(-1,1,order='F')
```

```python
Y_train.shape
```

```python
app=np.concatenate((Y_train,X_train),axis=1)
kp=pd.DataFrame(app)
p=kp.corr()
p[0]
```
