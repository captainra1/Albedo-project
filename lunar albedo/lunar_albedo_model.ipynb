## data preprocessing

```python
# now splitting the label dataset into training set and test set........
```

```python
#function for data splitting:
def data_splitter(data):
    F1=data.iloc[:,0:360].values
    F2=data.iloc[:,360:720].values
    return F1,F2
```

```python
Y_label_train,Y_label_test=data_splitter(albedo_map_data)
```

```python
K_features_train,K_features_test=data_splitter(K_map_data)
```

```python
Th_features_train,Th_features_test=data_splitter(Th_map_data)
```

```python
Ti_features_train,Ti_features_test=data_splitter(Ti_map_data)
```

```python
#  now, in order to make a feature set, we have to merge all the different chemical datasets into one with proper positon
# so that we can use them to train and predict.
```

```python
# way of merging different chemical datasets: first we will convert the all chemical dataset into a column vector after that
# we will merge all column vector column-wise.
# for label data set, we will also convert it into column vector.
```

```python

def column_converter(data):
    column_arr=data.reshape(-1,1,order='F') # here order='F' specifies that it will read column vice
    return column_arr
```

```python
Fe_column_train=column_converter(Fe_features_train)
```

```python
Fe_column_train.shape
```

```python
K_column_train=column_converter(K_features_train)

Th_column_train=column_converter(Th_features_train)

Ti_column_train=column_converter(Ti_features_train)
```

```python
# now all the data is converted to column vector now we will add them.
```

```python
X_train=np.concatenate((Fe_column_train,K_column_train,Th_column_train,Ti_column_train),axis=1)
```

```python
X_train.shape
```

```python
# now we have our feature set ready,now converting label set.
```

```python
Y_train=column_converter(Y_label_train)
```

```python
Y_train.shape
```

```python
# I have applied many feature engineering but none of them giving good result and getting even worse, so no feature
#engineering on dataset.
```

```python
# seeing corelation between target and featureset.
comb_data=np.concatenate((Y_train,X_train),axis=1)
kp=pd.DataFrame(comb_data)
p=kp.corr()
p[0]
```

##  Model Training.

```python
Lin_model=LinearRegression().fit(X_train,Y_train)
```

```python
DT_model=DecisionTreeRegressor().fit(X_train,Y_train)
```

```python
RF_model=RandomForestRegressor().fit(X_train,Y_train.ravel())
```

```python
Rd_model=Ridge().fit(X_train,Y_train) 
```

## checking the model credibility via cross velidation.

```python
from sklearn.model_selection import cross_val_score
```

```python
# function for cross validation score.
def crossval_score(model,X,Y):
    score=cross_val_score(model,X,Y,scoring='r2')
    return score.mean(),score.std()
```

```python
#dataset for cross validation.
X_cval=X_train[0:150,:]
Y_cval=Y_train[0:150,:]
```

```python
# Linear reg cross validation score
cv_mean,cv_std=crossval_score(Lin_model,X_cval,Y_cval)
print("mean:linear reg model:",cv_mean)
print("std:Linear reg model:",cv_std)
```

```python
# ridge cross validation score
cv2_mean,cv2_std=crossval_score(Rd_model,X_cval,Y_cval)
print("mean:ridge model:",cv2_mean)
print("std:ridge model:",cv2_std)
```

```python
# decision tree cross validation score
cv2_mean,cv2_std=crossval_score(DT_model,X_cval,Y_cval)
print("mean:decision tree model:",cv2_mean)
print("std:decision tree model:",cv2_std)
```

```python
# random forest cross validation score
cv2_mean,cv2_std=crossval_score(RF_model,X_cval,Y_cval)
print("mean:decision tree model:",cv2_mean)
print("std:decision tree model:",cv2_std)
```

```python
 # so, ridge model is performing best out of all algorithms.
```

## processing testing data.

```python
Fe_column_test=column_converter(Fe_features_test)

K_column_test=column_converter(K_features_test)

Th_column_test=column_converter(Th_features_test)

Ti_column_test=column_converter(Ti_features_test)

X_test=np.concatenate((Fe_column_test,K_column_test,Th_column_test,Ti_column_test),axis=1)
```

```python
predicted_albedo_data=Rd_model.predict(X_test)
```

```python
# converting to dimentions and plotting right side predicted albedo.
right_albedo_data=predicted_albedo_data.reshape(360,-1,order='F')             
plt.subplots(figsize=(8,7))
sns.heatmap(right_albedo_data,xticklabels=False,yticklabels=False,cmap='gray')
```

## checking performance.

```python
# model accuracy.
original_right_albedo=Y_label_test
Y_test=column_converter(Y_label_test)
r2_score(original_right_albedo,right_albedo_data)
```
